### 3.Plot a bar graph of male vs female buyers participated in the sales

```python
df['Buyer Gender'].value_counts()
```

```python
sns.countplot(df['Buyer Gender'])
plt.title('Male Vs Female buyers',fontsize=14)
plt.xlabel('Buyer Gender',fontsize=12)
plt.grid()
```

```python
#Alternative approach
```

```python
plt.figure(figsize=(8,6))
plt.hist(df["Buyer Gender"],color='red')
plt.title('Male Vs Female buyers',fontsize=16)
plt.xlabel("Buyer Gender",fontsize=14)
plt.ylabel('count',fontsize=14)
plt.grid(color='green')
```

### 4.Find the top 5 cars based on their sales price.

```python
top5=df.nlargest(5,'Sale Price')
top5=top5.reset_index()
```

```python
top5.head()
```

```python
top5.loc[:,['Make','Model','Sale Price']].head()
```

### 5. Find the least 5 cars based on their Resell price.

```python
least5=df.nsmallest(5,'Resell Price')
least5=least5.reset_index()
```
